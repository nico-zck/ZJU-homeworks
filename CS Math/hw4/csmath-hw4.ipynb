Implement the Levenberg-Marquardt method

1.sample the function curve of y=sin(x) with Gaussian noise

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def curve_func(x, params):
    a = params[0, 0]
    b = params[1, 0]
    c = params[2, 0]
    return a * np.sin(b * x) + b * np.cos(c * x)
```

```python

m = 100
true_params = np.mat([1, 3, 2]).T
curve_x = np.linspace(0, 5, m)
curve_y = curve_func(curve_x, true_params)
data_y = curve_func(curve_x, true_params) + np.random.normal(0, 0.5, m)

plt.plot(curve_x, curve_y)
plt.scatter(curve_x, data_y, marker='o', c='', edgecolors='g')
```

### Levenberg-Marquardt Algorithm

```python
def deriv(x, y, params, n):
    delta = 0.000001
    p1 = params.copy()
    p2 = params.copy()
    p1[n, 0] -= delta
    p2[n, 0] += delta
    f1 = curve_func(x, p1)
    f2 = curve_func(x, p2)
    r1 = y - f1
    r2 = y - f2
    d = (r2 - r1) / (2 * delta)
    return d


def jacobian_matrix(X, Y, params):
    m = X.shape[0]
    n = params.shape[0]
    J = np.mat(np.zeros((m, n)))
    for i in range(m):
        for j in range(n):
            J[i, j] = deriv(X[i], Y[i], params, j)
    return J


def LM(X, Y, lamda=0.01, max_iter = 100):
    # initialization for all parameters
    init_params = np.mat([0.6, 3.6, 1.5]).T
    v =2
    
    last_mse = 0

    for i in range(max_iter):
        fx = curve_func(X, init_params)
        r = Y - fx
        mse = np.mean(np.square(r))

        J = jacobian_matrix(X, Y, init_params)
        H = J.T * J + lamda * np.eye(init_params.shape[0])
        dx = -H.I * J.T * r[..., None]
        params_tmp = init_params.copy()
        params_tmp += dx

        fx_tmp = curve_func(X, params_tmp)
        r_tmp = Y - fx_tmp
        mse_tmp = np.mean(np.square(r_tmp))

        q = (mse - mse_tmp) / (0.5 * dx.T * (lamda * dx - J.T * r[..., None]))[0, 0]

        if q > 0:
            s = 1.0 / 3.0
            v = 2
            mse = mse_tmp
            init_params = params_tmp
            temp = 1 - pow(2 * q - 1, 3)

            if s > temp:
                lamda = lamda * s
            else:
                lamda = lamda * temp
        else:
            lamda = lamda * v
            v = 2 * v
            init_params = params_tmp

        print("step = %d, mse = %.8f" % (i, mse))

        if abs(mse - last_mse) < 0.0000001:
            break
        last_mse = mse
    return init_params
```

```python
params = LM(curve_x, data_y)
print('final parameters is :', params)
print('true parameters is: ', true_params)
pred_y = curve_func(curve_x, params)

plt.plot(curve_x, curve_y)
plt.scatter(curve_x, data_y, marker='o', c='', edgecolors='g')
plt.plot(curve_x, pred_y, 'r')
```
