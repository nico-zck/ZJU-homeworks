3.Estimate parameters of generated data via step2 by using the E-M method

#### 实现EM算法

```python
from scipy.stats import multivariate_normal

def phi(Y, mu_k, cov_k):
    '''
    第 k 个高斯模型在X上的概率值
    '''
    norm = multivariate_normal(mean=mu_k, cov=cov_k)
    return norm.pdf(Y)

def getExpectation(Y, mu, cov, alpha):
    '''
    E 步：计算每个模型对样本的响应度
    Y 为样本矩阵 N x D
    mean 为均值多维数组 K x D
    cov 为协方差矩阵的数组 K x D x D
    alpha 为模型响应度数组 K x 1
    '''
    # 样本数
    N = Y.shape[0]
    # 模型数
    K = alpha.shape[0]

    assert N > 1, "There must be more than one sample!"
    assert K > 1, "There must be more than one gaussian model!"

    # 响应度矩阵，行对应样本，列对应响应度
    gamma = np.mat(np.zeros((N, K)))

    # 计算各模型中所有样本出现的概率，行对应样本，列对应模型
    prob = np.zeros((N, K))
    for k in range(K):
        prob[:, k] = phi(Y, mu[k], cov[k])
    prob = np.mat(prob)

    # 计算每个模型对每个样本的响应度
    for k in range(K):
        gamma[:, k] = alpha[k] * prob[:, k]
    for i in range(N):
        gamma[i, :] /= np.sum(gamma[i, :])
    return gamma



def maximize(Y, gamma):
    '''
    M 步：迭代模型参数
    Y 为样本矩阵 N x D
    gamma 响应度矩阵 N x K
    '''
    # 样本数和特征数
    N, D = Y.shape
    # 模型数
    K = gamma.shape[1]

    #初始化参数值
    mu = np.zeros((K, D))
    cov = []
    alpha = np.zeros(K)

    # 更新每个模型的参数
    for k in range(K):
        # 第 k 个模型对所有样本的响应度之和
        Nk = np.sum(gamma[:, k])
        # 更新 mu 对每个特征求均值
        mu[k, :] = np.sum(np.multiply(gamma[:, k], Y), axis=0) / Nk
        # 更新 cov
        cov_k = np.zeros((D, D))
        for i in range(N):
            dev = np.mat(Y[i]- mu[k])
            cov_k += gamma[i, k] * dev.T * dev / Nk
        cov.append(cov_k)
        # 更新 alpha
        alpha[k] = Nk / N
    cov = np.array(cov)
    return mu, cov, alpha
```

### 给高斯分布加上范围

```python
from matplotlib.patches import Ellipse
def plot_ellipse(mean, cov):
    '''
    将高斯分布以椭圆的形式画出来
    '''
    for mu, sigma in zip(mean, cov):
        vals, vecs = np.linalg.eigh(sigma)
        theta = np.degrees(np.arctan2(*vecs[:,1][::-1]))
        height, width = 4 * np.sqrt(vals)
        ellip = Ellipse(xy=mu, width=width, height=height, angle=theta, linewidth=2, fill=False)
        ax = plt.gca()
        ax.add_artist(ellip)
    plt.axis('equal')
    plt.show()
plt.scatter(x1, y1)
plot_ellipse([mean1], [cov1])
```

### 对高斯混合模型使用EM算法对生成的样本进行估计

```python
# 高斯模型数量
K = 2
# 迭代次数
epoch = 10

data = np.stack([x,y]).T
N, D = data.shape

mean = np.random.rand(K, D)
cov = [np.eye(D)] * K
alpha = np.array([1.0 / K] * K)

%precision 2
for i in range(epoch):
    gamma = getExpectation(data, mean, cov, alpha)
    mean, cov, alpha = maximize(data, gamma)
    if i%2 == 0:
        print("epoch: %d" % i, "mean:", mean, "cov:", cov, "alpha:", alpha, sep='\n', end='\n------------------\n')
        plt.scatter(x, y)
        plot_ellipse(mean, cov)
print("{sep} Result {sep}".format(sep="-" * 20))
print("mean:", mean, "cov:", cov, "alpha:", alpha, sep="\n")
```
